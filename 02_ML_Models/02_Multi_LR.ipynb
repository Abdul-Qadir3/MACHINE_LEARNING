# To understand test and split

```python
# importing the necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# splitting the data set into X and y 
X=df[["sepal_length","sepal_width"]]
y=df["petal_width"]

# splitting the data set into training and into testing data
# random stat in which same patch of the data will be 80# else it will pick rondomlly whicchanges the result
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)

# training the model on training dataset
model = LinearRegression()
model.fit(X_train,y_train)

#prdicting the values of testing dataset
y_pred = model.predict(X_test)

#print the slope and intercept of regression line
print('intercept: ', model.intercept_)
print('slope: ',model.coef_)
```

```python
# plotting scattered plot on the model
# Importing the matplotlib library, specifically the pyplot module, which provides functions for creating plots and visualizations.
import matplotlib.pyplot as plt

# Creating a scatter plot to compare the actual values (y_test) with the predicted values (y_pred).
# y_test: The actual values of the dependent variable.
# y_pred: The predicted values of the dependent variable.
plt.scatter(y_test, y_pred)

# Adding labels to the x-axis and y-axis of the plot.
plt.xlabel('Actual values')
plt.ylabel('Predicted Values')

# Adding a title to the plot.
plt.title('Actual vs Predicted Values')

# Displaying the plot.
plt.show()
```

```python
# Using the metrics for evaluating the model
print("R2 score: ", r2_score(y_test,y_pred))
print("Mean Absolute Error: ", mean_absolute_error(y_test, y_pred))
print("Mean Square Error: ", mean_squared_error(y_test,y_pred))
```

# What happens when R2 is high
A high R2 value indicates that the linear regression `model fits the data well` and can explain a large proportion of the variability in the response variable. Specifically, an R2 value `close to 1 indicates` that the model can explain most of the variation in the response variable, while an R2 value `close to 0 indicates` that the model does not explain much of the variation in the response variable.

1.	`Strong Relationship:` A high R2 value suggests that there is a `strong relationship between the independent and dependent variables`. In other words, changes in the independent variable(s) are associated with predictable changes in the dependent variable.
2.	`Good Fit:` The regression line (or plane in the case of multiple independent variables) fits the data points closely. This means that the model is able to `accurately predict the values of the dependent variable based on the values of the independent variable(s)`.
3.	`Less Variability:` A `high R2 value indicates that less variability` in the dependent variable is left unexplained by the independent variable(s). This suggests that there are fewer random fluctuations or noise in the relationship between the variables.
4.	`Confidence in Predictions:` With a high R2, we can have more confidence in the predictions made by the regression model. It implies that the model is reliable and can be used to make accurate predictions about the dependent variable based on the independent variable(s).

# What happens when MAE and MSE is high?
- `A high Mean Absolute Error (MAE) or Mean Squared Error (MSE) value indicates that the linear regression model is not fiting the data well`
- In the case of MAE, a high value means that the model's predicted values are, on average, far from the actual values. In other words, the model's predictions have a high degree of error.

- Similarly, in the case of MSE, a high value means that the squared difference between the model's predicted values and the actual values is large.
This indicates that the model's predictions are not accurate and have a high degree of error.
- In summary, a high MAE or MSE valie suggests that the linear regression model may not be the best fit for the data, and there may be other
models or techniques that can provide better predictions.
