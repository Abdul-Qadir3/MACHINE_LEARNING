# Simple linear regreesion 
Simple inear regression model is a statistical method that allows to summarize and study the relationship between two continous (quantitative ) variables. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ML library
from sklearn.linear_model import LinearRegression
```

```python
#step 1: import dataset and load dataset
# Loading a dataset named 'tips' using the load_dataset() function from seaborn.
# This dataset contains information about tips in a restaurant.
# It's a built-in dataset provided by seaborn for demonstration purposes.
df = sns.load_dataset('tips')

# Selecting the 'total_bill' column from the DataFrame 'df' and assigning it to the variable X.
# This column represents the total bill amount for each transaction.
X = df[['total_bill']]

# Selecting the 'tip' column from the DataFrame 'df' and assigning it to the variable y.
# This column represents the tip amount for each transaction, which is the target variable we want to predict.
y = df['tip']
```

```python
#step 2: create model
# Creating an instance of the LinearRegression class
model = LinearRegression()
```

```python
#step 3: fit the model(train the model)
# Fitting the model to the training data
# X: array-like, shape (n_samples, n_features)
# y: array-like, shape (n_samples,)
# This step trains the model by finding the coefficients that minimize the residual sum of squares between the observed and predicted target values.
model.fit(X, y)
```

```python
#step 4: make predictions
# Making predictions using the trained model
# Predictions can be made on new data using the predict method
# Here, we're passing a new feature value (12) to predict the target value
# The input to predict must be a 2D array, so we pass [[12]]
# Returns the predicted target value(s) for the given feature value(s)
predicted_value = model.predict([[12]])
```

```python
# plot the model
# Creating a scatter plot using Seaborn
# The scatterplot function is used to create a scatter plot between 'total_bill' and 'tip' columns in the DataFrame 'df'
sns.scatterplot(x="total_bill", y="tip", data=df)

# Adding labels to the x-axis and y-axis
plt.xlabel('total_bill')  # Label for the x-axis
plt.ylabel('tip')  # Label for the y-axis

# Displaying the plot
plt.show()  # This command displays the plot created using Matplotlib
```

```python
# plot the model
# Plotting a scatter plot using seaborn
# 'total_bill' on the x-axis, 'tip' on the y-axis, and data from the DataFrame df
sns.scatterplot(x="total_bill", y="tip", data=df)

# Plotting the best-fitted line of the model for logistic regression
# X is the input variable used for prediction
# model.predict(X) generates the predicted values based on the model
# color='red' specifies the color of the line
plt.plot(X, model.predict(X), color="red")

# Adding labels to the axes
plt.xlabel('total_bill')  # Label for the x-axis
plt.ylabel('tip')         # Label for the y-axis

# Displaying the plot
plt.show()
```

```python
# slope , intercept and r^2 value
# Print the intercept of the linear regression model.
print("intercept: ", model.intercept_)

# Print the coefficients (slopes) of the features in the linear regression model.
print("Slope: ", model.coef_)

# Print the coefficient of determination (R^2) of the linear regression model.
# The coefficient of determination measures the goodness of fit of the model,
# indicating how well the independent variables explain the variability of the dependent variable.
print("R^2: ", model.score(X, y))
```


`Ordinary Least Squares (OLS) regression is a statistical method used to model the relationship between one or more independent variables and a dependent variable. It is a type of linear regression analysis where the goal is to find the best-fitting line that minimizes the sum of the squared differences between the observed and predicted values of the dependent variable.`

In OLS regression:

The dependent variable (also known as the response variable) is denoted as y.
The independent variable(s) (also known as predictor variable(s)) are denoted as X.
The relationship between the dependent variable y and the independent variable(s) X is assumed to be linear.
The OLS regression model can be represented as:




```python
# Importing the statsmodels library, which provides classes and functions for statistical modeling and analysis.
import statsmodels.api as sm

# Creating a model using Ordinary Least Squares (OLS) regression.
# y: dependent variable (response variable)
# X: independent variable(s) (predictor variable(s))
# sm.add_constant(X): Adding a constant term to the independent variable(s) for the intercept in the regression model.
# fit(): Fitting the model to the data and estimating the parameters.
model = sm.OLS(y, sm.add_constant(X)).fit()

# Getting the p-value of the coefficient associated with the independent variable(s) in the regression model.
# model.params[1]: Accessing the parameter estimate of the independent variable(s).
# p-value represents the probability that the coefficient is equal to zero (indicating no effect).
p_value = model.params[1]

# Printing the p-value.
print("p-value: ", p_value)
```

# SLR on scalled data

```python
# Importing necessary libraries for data analysis and visualization
import pandas as pd          # Pandas for data manipulation
import numpy as np           # NumPy for numerical operations
import matplotlib.pyplot as plt  # Matplotlib for plotting
import seaborn as sns        # Seaborn for statistical visualization

# Importing machine learning libraries
from sklearn.linear_model import LinearRegression   # Linear Regression model
from sklearn.preprocessing import StandardScaler, MinMaxScaler  # Scalers for data normalization

# Loading a sample dataset from Seaborn for demonstration
df = sns.load_dataset("tips")  # Loading the 'tips' dataset

# Extracting independent variable (features) 'total_bill' and dependent variable (target) 'tip'
X = df[['total_bill']]   # Independent variable (features)
y = df['tip']            # Dependent variable (target)

# Initializing a scaler for data scaling (normalization)
# Choose between MinMaxScaler or StandardScaler based on requirement
scaler = MinMaxScaler()   # Using MinMaxScaler for scaling the data
# scaler = StandardScaler()  # Using StandardScaler for scaling the data

# Scaling the independent variable 'X'
X_scaled = scaler.fit_transform(X)   # Fit the scaler to the data and transform it

# Creating a Linear Regression model
model = LinearRegression()   # Initializing a Linear Regression model

# Training (fitting) the Linear Regression model on the scaled data
model.fit(X_scaled, y)   # Training the model on the scaled independent variable 'X_scaled' and dependent variable 'y'
```

```python
# slope , intercept and r^2 value for standard scaler
# Printing the intercept term of the linear regression model.
# model.intercept_: Accessing the intercept term of the fitted model.
print("intercept: ", model.intercept_)

# Printing the coefficients (slopes) of the independent variables in the linear regression model.
# model.coef_: Accessing the coefficients (slopes) of the independent variables.
print("Slope: ", model.coef_)

# Printing the R-squared value of the linear regression model.
# R-squared (R^2) represents the proportion of variance in the dependent variable that is explained by the independent variables.
# model.score(X_scaled, y): Computing the R-squared value of the model based on the scaled independent variables (X_scaled) and the dependent variable (y).
print("R^2: ", model.score(X_scaled, y))
```

```python
# slope , intercept and r^2 value for MinMax scaler
print("intercept: ", model.intercept_)
print("Slope: ", model.coef_)
print("R^2: ", model.score(X_scaled,y))
```
